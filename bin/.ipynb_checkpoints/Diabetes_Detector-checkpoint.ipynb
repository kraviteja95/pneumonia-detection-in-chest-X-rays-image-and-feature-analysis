**Additional Details about the Variables:**

***Target Variable:***

- **Diabetes_binary** -> 0 = no diabetes | 1 = prediabetes or diabetes

***Input Features:***

- **HighBP** -> `0 = no high BP | 1 = high BP`
- **HighChol** -> `0 = no high cholesterol | 1 = high cholesterol`
- **CholCheck** -> `0 = no cholesterol check in 5 years | 1 = yes cholesterol check in 5 years`
- **BMI** -> Body Mass Index
- **Smoker** -> Have you smoked at least 100 cigarettes in your entire life? Note: 5 packs = 100 cigarettes -> `0 = no | 1 = yes `
- **Stroke** -> (Ever told) you had a stroke. `0 = no | 1 = yes`

- **Veggies** -> Consume Vegetables 1 or more times per day `0 = no 1 = yes`
- **HvyAlcoholConsump** -> Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week) `0 = no | 1 = yes`
- **AnyHealthcare** -> Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. `0 = no | 1 = yes`
- **NoDocbcCost** -> Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? `0 = no | 1 = yes`
- **GenHlth** -> Would you say that in general your health is: scale 1-5 `1 = excellent | 2 = very good  | 3 = good | 4 = fair | 5 = poor`
- **MentHlth** -> Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good? scale `1-30` days
- **PhysHlth** -> Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good? scale `1-30` days
- **DiffWalk** -> Do you have serious difficulty walking or climbing stairs? `0 = no | 1 = yes`
- **Sex** -> `0 = female | 1 = male`
- **Age** -> 13-level age category (_AGEG5YR see codebook) `1 = 18-24 | 9 = 60-64 | 13 = 80 or older`
- **Education** -> Education level (EDUCA see codebook) scale `1-6 | 1 = Never attended school or only kindergarten -> 2 = Grades 1 through 8 (Elementary) 3 = Grades 9 through 11 (Some high school) 4 = Grade 12 or GED (High school graduate) 5 = College 1 year to 3 years (Some college or technical school) 6 = College 4 years or more (College graduate)`
- **Income** -> Income scale (INCOME2 see codebook) scale `1-8 | 1 = less than $10,000 | 5 = less than $35,000 | 8 = $75,000 or more`

$$ z = \frac{x - \mu}{\sigma} $$

Here, from the plots noticed above, the distribution is almost or very near to the normal for the `BMI` feature. The other features have skewness. To detect and remove the outliers using z-score, we have to consider the feature that is almost normal.

## Bivariate Analysis

We have 2 items to understand here.

- **Association** measures the strength of the relationship between variables, without implying causation or direction.
- **Correlation** measures the strength and direction of the linear relationship between continuous variables.

### Chi-Square Test for Binary Categorical Variables

This statistical test is employed to identify whether there is a significant association between two categorical variables. It quantifies the strength of the relationship between these variables, but it does not imply causality.

Bivariate analysis compares two variables, which can be thought of as input vs input variables. 

To do this for binary categorical variables, we can calculate the chi-squared statistic, degree of freedom and p-value of each variable.

This should be done by performing cross tab of each variable.

Chi-Squared Formula:

$$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

Degree of Freedom Formula:

$$ \text{DF} = (r - 1)(c - 1) $$

Goodness of Fit Test Formula:

$$ \text{DF} = k - 1 $$

Independence Test Formula:

$$ \text{DF} = (r - 1)(k - 1) $$

Where:
- $O_i$ is the observed frequency
- $E_i$ is the expected frequency
- $r$ is the number of rows in the contingency table
- $c$ is the number of columns in the contingency table
- $k$ is the number of categories

If you look at the below table, all the binary categorical variable have p-value to be 0.0000 which is less than 0.05 which means all the features have strong impact on the target variable.
The strongest impact is `HighBP` feature.

````python
chi2_results_df = eda.chi_squared_test_for_categorical_variables(diabetes_df, binary_columns)
chi2_results_df
````

## Mutual Information for Numerical Categorical Variables

This is a measure of the mutual dependence between two variables. It quantifies the amount of information that one variable contains about another variable. Mutual information is a measure of association, not correlation.

Mutual Information is calculated to the features which are of ordinal categorical type. Look at the formula of the Mutual Information below.

````python
from IPython.display import Math

Math(r'$I(X;Y) = \sum_{x \in X} \sum_{y \in Y} p(x,y) \log_2 \frac{p(x,y)}{p(x)p(y)}$')
````

Where:
- $I(X;Y)$ is the mutual information between the random variables $X$ and $Y$
- $p(x,y)$ is the joint probability distribution of $X$ and $Y$
- $p(x)$ is the marginal probability distribution of $X$
- $p(y)$ is the marginal probability distribution of $Y$

````python
final_mutual_info = eda.mutual_info_score_for_ordinal_categorical_variables(resampled_diabetes_df)
final_mutual_info
````

The Mutual information results usually range from 0 to a maximum value that depends on the entropy of the variables. This indicates the stronger relationship between the variables.

## Pearson Correlation for Binary Categorical Variables

Pearson correlation is also one of the measures to calculate the strong correlation between each variable vs each variable.

Formula to do it is:

$$ r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2 \sum_{i=1}^n (y_i - \bar{y})^2}} $$

where:

- $r$ is the Pearson correlation coefficient.
- $x_i$ and $y_i$ are the ith values of the two variables.
- $\bar{x}$ and $\bar{y}$ are the means of 2 variables.
- $n$ is the number of data points.

````python
# Calculate Pearson correlation coefficient between each feature and the target variable
pearson_correlations = eda.pearson_correlation_for_binary_categorical_variables(
    resampled_diabetes_df)

# Print the correlation coefficients
pearson_correlations
````

**Formulas used:**

**Precision** is a metric that measures the accuracy of a model's positive predictions. It is calculated as:

$$ \text{Precision} = \frac{\text{TP}}{\text{TP} + \text{FP}} $$

**Recall** is a metric that assesses the completeness of a model's positive predictions. It is calculated by dividing the number of true positive predictions (TP) by the sum of true positives and false negatives (FN).

$$ \text{Recall} = \frac{\text{TP}}{\text{TP} + \text{FN}} $$

**F1-Score:** is a weighted average of precision and recall, calculated as the harmonic mean of the two. This metric aims to strike a balance between precision and recall, providing a more comprehensive evaluation than either metric alone.

$$ \text{F1-score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} $$

**Support:** refers to the number of instances of a specific class within the actual data. This value is crucial in multi-class classification problems as it is used to compute the precision, recall, and F1-score for each class.

$$ \text{Support} = \frac{\text{Number of samples of the given class}}{100} $$
